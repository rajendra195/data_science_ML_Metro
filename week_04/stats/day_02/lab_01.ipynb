```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import statistics # inbuild python module
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df["total_bill"].mean()
```

```python
df["total_bill"].median()
```

```python
statistics.mode(df["total_bill"])
```

```python
df["total_bill"].unique()
```

```python
sns.boxplot(df["total_bill"])
sns.stripplot(df["total_bill"])
```

```python
sns.histplot(df["total_bill"], kde=True, bins=30)
plt.axvline(np.mean(df["total_bill"]), color="Red")
plt.axvline(np.median(df["total_bill"]), color="Green", linestyle=)
plt.axvline(statistics.mode(df["total_bill"]), color="Purple")
```
