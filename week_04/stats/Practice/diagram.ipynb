```python
cars = df["Manufacturer"].unique()
print(cars)
```

```python
counts = []
for i in cars:
    con = df[df["Manufacturer"] == i].Manufacturer.count()
    counts.append(con)
counts
```

```python
plt.figure(figsize=(8, 6))
plt.bar(x=cars, height=counts, color="purple")
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=cars, y=counts, hue=["red", "purple", "skyblue", "orange", "black", "yellow", "green", "blue", "scarlet"])
plt.xlabel('Cars')
plt.ylabel("Counts")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(counts, labels=cars, autopct="%.2f%%")
plt.show()
```
