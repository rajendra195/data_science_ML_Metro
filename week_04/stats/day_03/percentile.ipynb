```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Consider datasets
data = [2, 2, 3, 4, 5, 5, 5, 6, 7, 8, 8, 8, 8, 8, 9, 9, 10, 11, 11, 12]
```

```python
# sort the data
sorted_data = sorted(data)
print(sorted_data)
```

```python
def compute_percentile(sorted_data, value):
    data_len_below_value = len(sorted_data[:sorted_data.index(value)])
    n = len(sorted_data)
    print(data_len_below_value)
    percentile_rank = data_len_below_value / n
    
    return percentile_rank

percentile_rank_of_10 = compute_percentile(sorted_data=sorted_data, value=10)
print(f"Percentile rank of 10 is: {percentile_rank_of_10}")
```

# Boxplot

```python
data = [2, 2, 3, 4, 5, 5, 5, 6, 7, 8, 8, 8, 8, 8, 9, 9, 10, 11, 11, 12]
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, color="Purple")
sns.stripplot(data=data)
plt.xlabel("Data")
plt.ylabel("Values")
plt.title("Box plot of Data points")
plt.show()
```

```python
data = [-20, 2, 3, 4, 5, 5, 5, 6, 7, 8, 8, 8, 8, 8, 9, 9, 10, 11, 11, 12, 50] # gives outliers
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, color="Purple")
sns.stripplot(data=data)
plt.xlabel("Data")
plt.ylabel("Values")
plt.title("Box plot of Data points")
plt.show()
```
