# 1. Line plot

* Line plots are used to visualize the relationship between two variables over a continuous interval. 
* They are suitable for showing trends and patterns in data, such as changes over time or across different conditions.

```python
import pandas as pd
import matplotlib.pyplot as plt

score_df = pd.read_csv("csv_data/cricket_scores.csv")
score_df.head()
```

```python
score_df.Batting_team.unique()
```

```python
score_nz = score_df[score_df["Batting_team"]=="Newzeland"]
score_ind = score_df[score_df["Batting_team"] == "India"]
```

```python
plt.plot(score_nz["Over"], score_nz["Run"], marker="o", color="red", label="New Zealand")
plt.plot(score_ind["Over"], score_ind["Run"], marker="o", color="green", label="India")

plt.xlabel("Overs")
plt.ylabel("Scores")
plt.title("New Zealand vs India t20 Scores")
plt.legend()

plt.show()
```

* The x-axis of the above figure shows the total overs of the game whereas the y_axis shows the scores scored by the team.
* The red line shows the score scored by Newzealand team and green line shows the score scored by Indian team

# 2. Bar Graph
#### Use Cases:* Bar charts are ideal for comparing values across distinct categories. They excel at highlighting differences between groups or showcasing the composition of a whole.
  
* They are best suited for discrete data, where categories are distinct and have no inherent order (e.g., types of fruits, customer demographics, product sales by category).
* While not their primary strength, bar charts can be used to show trends over time with discrete intervals (e.g., monthly sales figures, website traffic by day).ay).y).

```python
company_df = pd.read_csv("csv_data/Fortune_1000.csv")
company_df.head()
```

```python
sector_df = company_df['sector'].value_counts().to_frame().reset_index()
sector_df.columns = ['sector', 'frequency']
sector_df.head()
```

```python
plt.figure(figsize = (30, 10))
plt.bar(sector_df['sector'], sector_df['frequency'])

plt.title('Sectorwise Distribution of Fortune 1000 Companies')
plt.xlabel('Sectors')
plt.ylabel('No of companies')

plt.xticks(rotation = 45)

plt.show()
```

* Above diagram shows the number of companies of each sectors in the bar graph.
* It is easy to visualize each sectors and their numbers using the bar plot.
* Here, highest number company is in financials sectors.

# 3. Histogram
### Use Cases:
* Histograms visualize the frequency distribution of continuous data, revealing the shape (normal, skewed, etc.) and spread of the data.
  
* They help identify patterns like central tendency (mean, median), outliers, and skewness in the data.
  
* Histograms are useful for understanding the variability within a dataset.

```python
hr_df = pd.read_csv("csv_data/hr_comma_sep.csv")
hr_df.head(2)
```

```python
mean = hr_df["satisfaction_level"].mean()
median = hr_df["satisfaction_level"].median()
mode = hr_df["satisfaction_level"].mode()


plt.figure(figsize=(8, 6))
plt.hist(hr_df["satisfaction_level"],  color="green", edgecolor="black")
plt.axvline(mean, color="orange")
plt.axvline(median, color="red", linestyle="--")
plt.axvline(mode[0])
plt.xlabel("Satisfaction Level of the employees")
plt.ylabel("Number of the employees (frequency)")
plt.show()
```

* Above histogram plot shows it is left skewed which can also be concluded from mean and median because median is greater than mean.* 
In the above figure we can see that most of the employees are satisfied with their job as their satisfaciton level is lies between 0.4 to 1.

# 4. Pie Chart
A circular chart divided into slices, where each slice represents a category and its area is proportional to the category's value compared to the whole.
### Use cases:
* Ideal for displaying the composition of a whole, highlighting the relative sizes of different categories.
  * 
Effective for comparing a few distinct categories (typically 4-6) and their contribution to the total

* 
Not ideal for comparing many categories or showing trends over time.

```python
def freq_count_df(df, col_name, renamed_cols):
    new_df = df[col_name].value_counts().to_frame().reset_index()
    new_df.columns = renamed_cols
    return new_df

car_df = pd.read_csv('csv_data/car_details.csv')
fuel_df = freq_count_df(car_df, 'fuel', ['fuel', 'frequency'])
fuel_df.head()
```

```python
plt.pie(x = fuel_df['frequency'])

plt.legend(fuel_df['fuel'])
plt.show()
```

* Above Piechart shows the frequency of fuels.

```python
expenses = [30000, 20000, 15000, 10000, 5000]
categories = ['Rent', 'Salaries', 'Utilities', 'Marketing', 'Miscellaneous']

plt.pie(expenses, labels=categories, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon', 'lightpink'])
plt.title('Annual Expenses Distribution')

plt.show()
```

* Each Slices of the pie represent different expense categories. Each slice's size corresponds to the proportion of the total expenses attributed to that category.

# 5. ScatterPlot
A graph where each data point is represented by a symbol (dot, circle, etc.) plotted on horizontal and vertical axes.

### Use cases:
* Useful for exploring relationships between two continuous variables.
  * 
Can reveal trends, correlations, or clusters between the variables

* 
Can identify outliers that might influence the analysis.

```python
plt.figure(figsize = (8,6))
plt.scatter(x = car_df['km_driven'], y = car_df['selling_price'])
plt.title("Km Driven Vs Selling Price")
plt.xlabel("Km Driven")
plt.ylabel("Selling Price")
plt.ticklabel_format(style = 'plain')
plt.show()
```

* Above diagram shows the scatter plot of the km driven of the car vs selling price of the cars.

```python
car_df['brand'] = car_df['name'].str.strip().str.split(' ').str[0].str.strip()
```

```python
tata_df = car_df.loc[car_df['brand'] == "Tata"]
```

```python
plt.figure(figsize = (10,8))
plt.scatter(x = tata_df['km_driven'], y = tata_df['selling_price'])
plt.title("Km Driven Vs Selling Price For Tata Cars")
plt.xlabel("Km Driven")
plt.ylabel("Selling Price")
plt.ticklabel_format(style = 'plain')
plt.show()
```

* Above diagram shows the scatter plot of the km drive vs selling price for only tata cars.

# 6. Box Plot
A rectangular box with a line dividing it in the middle (median). The box contains the middle 50% of the data (interquartile range). Lines extend from the box (whiskers) to show the range of data points within 1.5 times the interquartile range. Outliers are plotted as individual points beyond the whiskers.
### Use cases:
* Effective for comparing the distribution of data across different categories or groups.
  
* Provides insights into central tendency (median), variability (interquartile range), and potential outliers in the data.

```python
plt.boxplot(car_df['km_driven'])
plt.show()
```

* In the above diagram the middle orange line is median, just below the median is first_quartile and just above the median is third quartile.
* The buttom line below the first quartile is lower fench and line above third quartile is upper fench.
* All the black dot in the diagram are outliers.

```python
plt.boxplot(car_df['km_driven'])
plt.show()
```

# 7. Distribution Plot
This term can be ambiguous and refer to different plots depending on the context. Common examples include:
### a. Histogram: 
* Similar to a bar chart but used for continuous data. It displays the frequency distribution of data points across intervals (bins). Useful for understanding the shape (normal, skewed, etc.) and spread of the data.
### b. Kernel Density Plot: 
* A smoother version of a histogram, visualizing the probability density of the data. Offers a clearer picture of the underlying distribution compared to the binned approach of histograms.

### Use cases

**Histogram:** 
*  Understanding the distribution of continuous data like income levels, test scores, or website traffic - Identifying patterns like central tendency (mean, median), skewness, and outlier
s** **Kernel Density Pl**
* ** Similar uses to histograms but with a smoother representation of the data distribution.

```python
import seaborn as sns
import numpy as np

# Generating random heights for 1000 individuals
heights = np.random.normal(170, 10, 1000)  
sns.histplot(heights, kde=True, color='skyblue')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Distribution of Heights')
plt.grid(True)
plt.show()
```

```python
sns.kdeplot(heights, fill=True)
plt.axvline(np.mean(heights), color="red", linestyle="--")
plt.axvline(np.median(heights), color="black", linestyle="--")
plt.show()
```
