# 1. Simple Linear regression (y = 2x + 3)

```python
import matplotlib.pyplot as plt
import numpy as np

# x data from 100 to 200
x_data = np.arange(100, 200)

# respective y data with the value of x
y_data = [2 * x + 3 for x in x_data]

# plotting x data and y data
plt.figure(figsize=(10, 8))
plt.plot(x_data, y_data) 

plt.xlabel("x data")
plt.ylabel("y data")
plt.title("Simple linear regression of 'y = 2x + 3'")
# plt.scatter(x_data, y_data)

plt.grid()
plt.show()
```

# 4. Multiple Polynomial Regression (y = 2$x^3$ + 3$x^2$ + 4x + 5)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generating random values for x1, x2, and x3
num_points = 100
x1 = np.random.uniform(low=-10, high=10, size=num_points)
x2 = np.random.uniform(low=-10, high=10, size=num_points)
x3 = np.random.uniform(low=-10, high=10, size=num_points)

# Calculating the values of the expression for each combination of x1, x2, and x3
y = 3*x1 + 4*x2 + 5*x3 + 5

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, x3, c=y, cmap='viridis')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')
ax.set_title('Multiple Linear Expression: 3x1 + 4x2 + 5x3 + 5')
plt.show()
```
