# 1. Simple Linear regression (y = 2x + 3)

```python
import matplotlib.pyplot as plt
import numpy as np

# x data from 100 to 200
x_data = np.arange(100, 200)

# respective y data with the value of x
y_data = [2 * x + 3 for x in x_data]

# plotting x data and y data
plt.figure(figsize=(10, 8))
plt.plot(x_data, y_data) 

plt.xlabel("x data")
plt.ylabel("y data")
plt.title("Simple linear regression of 'y = 2x + 3'")
# plt.scatter(x_data, y_data)

plt.grid()
plt.show()
```

# 4. Multiple Polynomial Regression (y = 2$x^3$ + 3$x^2$ + 4x + 5)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])  # Features
y = np.array([2, 3, 4, 5])                      # Target

# Fit multiple linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict values
y_pred = model.predict(X)

# Plot actual data points
plt.scatter(X[:, 0], y, color='blue', label='Actual Data')
plt.scatter(X[:, 1], y, color='blue')

# Plot regression line
plt.plot(X[:, 0], y_pred, color='red', label='Regression Line')
plt.plot(X[:, 1], y_pred, color='red')

# Set labels and title
plt.xlabel('Feature 1')
plt.ylabel('Target')
plt.title('Multiple Linear Regression')
plt.legend()

# Show plot
plt.show()
```
