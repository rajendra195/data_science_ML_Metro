# Loading dataset and converting into list of lists

```python
import csv
with open("car_details.csv") as csvfile:
    data = csv.reader(csvfile)
    next(data)
    list_data = [item for item in data]
```

```python
list_data
```

# Creating a 2D numpy array from the list of lists

```python
import numpy as np
arr = np.array(list_data)
```
